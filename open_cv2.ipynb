```python
'''
Following code takes input the train image folder creates directory of sorted images by label
Then augment them and stores image in same breed folder, to create bottlenecks

'''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from os import listdir
from os.path import isfile, join
import os
import tensorflow as tf
import glob
```

```python
# Reading labels
labels = pd.read_csv('/home/hitesh/Downloads/capstone/labels.csv')
labels.head()
```

```python
## Putting images in sorted folder
for index, row in labels.iterrows():
    if not tf.gfile.Exists('/home/hitesh/Downloads/capstone/train_divided/%s'%(row['breed'])):
        tf.gfile.MkDir('/home/hitesh/Downloads/capstone/train_divided/%s'%(row['breed']))
        print('train/%s.jpg'%(row['id']),'dog_train/%s'%(row['breed']))
    tf.gfile.Copy('/home/hitesh/Downloads/capstone/train/%s.jpg'%(row['id']),'/home/hitesh/Downloads/capstone/train_divided/%s/%s.jpg'%(row['breed'],row['id']),True)
```

```python
### Augmentation code
def augmentation(image):
    
    
    
    
    img = cv2.imread(image)  #image read
    
    
    resized  = cv2.resize(img, (224,224))   #resize operation *224 selected for inception input
    rows,cols,channels = resized.shape
    
    #rotation 22,45,-22,-45 degree
    rows,cols,channels = resized.shape
    
    M_45 = cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
    
    M_315 = cv2.getRotationMatrix2D((cols/2,rows/2),315,1)
    
   
    rot_45 = cv2.warpAffine(resized,M_45,(cols,rows))
    
    rot_315 = cv2.warpAffine(resized,M_315,(cols,rows))
    
    #translation left, right, up,down
    
    M_minus100 = np.float32([[1,0,-25],[0,1,0]])  #Left shift by 100
   
    M_up100 = np.float32([[1,0,0],[0,1,-25]])   #up shift by 100
    
    translate_minus100 = cv2.warpAffine(resized,M_minus100,(cols,rows))
    
    translate_up100 = cv2.warpAffine(resized,M_up100,(cols,rows))
    
    
    #colorspace change to HSV
    hsv= cv2.cvtColor( resized, cv2.COLOR_BGR2HSV)
    
    
    os.system('rm '+image) #removing original image
    
    
    '''
    There are memory issues,so couple of augmentations not performed
    Apart from these rotations there are other couple of augmentations that we could have performed like 
    pulling out dog image from corner using both x &y translation
    
    other colorspace change like grayscale, wrap perspective
    
    
    '''
    
    
    
    a = image[:-4]
    
    ##saving images in same folder
    cv2.imwrite(image,resized)
    
    cv2.imwrite(a+'_rot_45.jpg',rot_45)
    cv2.imwrite(a+'_rot_315.jpg',rot_315)
    cv2.imwrite(a+'_trans_minus.jpg',translate_minus100)
    cv2.imwrite(a+'_trans_up.jpg',translate_up100)
    cv2.imwrite(a+'hsv.jpg',hsv)
    
    
    
    
    

    
    
    
    
    
    
    
    
    
    
    
```

```python
####saving augmented images in same folder
image_list = []
for filename in glob.iglob('/home/hitesh/Downloads/capstone/train_divided/**/*.jpg'):
    image_list.append(filename)
    

### augmenting images

for i in image_list:
    augmentation(i)
    


   
   
```
